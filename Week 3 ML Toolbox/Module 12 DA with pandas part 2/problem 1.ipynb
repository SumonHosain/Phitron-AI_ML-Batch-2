```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

print(df)

df=df.copy()
df['A'].fillna(0,inplace=True)
df['B'].fillna(99,inplace=True)
df['C'].fillna('Unknown',inplace=True)
print(df)

df = df.copy()
df.iloc[0:4,1] = 5
df.iloc[2,0] = np.nan
df.iloc[3,2] = None
print(df)

df = df.copy()
df['A'] = df['A'].fillna(3)
df.iloc[0,1] = 5
df.iloc[1:4,1] = 7.5
df['C'].fillna('Unknown',inplace=True)
print(df)
```
