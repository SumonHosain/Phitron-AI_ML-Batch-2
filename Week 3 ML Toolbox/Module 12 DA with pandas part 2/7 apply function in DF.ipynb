```python
import pandas as pd
df = pd.read_csv('new_Data.csv')
df['Data Structure Marks'].fillna(df['Data Structure Marks'].mean(),inplace=True)
df['Algorithm Marks'].fillna(df['Algorithm Marks'].mean(),inplace=True)
df['Python Marks'].fillna(df['Python Marks'].mean(),inplace=True)
df['Total Marks'] = df[['Algorithm Marks','Data Structure Marks','Python Marks']].sum(axis=1)
```

```python
# min max scaling
min = df['Total Marks'].min()
max = df['Total Marks'].max()
df['Scaled_values'] = df['Total Marks'].apply(lambda x:(x-min)/(max-min))
```

```python
#
def grading(marks):
    if marks >=260:
        return 'A+'
    elif marks >=250:
        return 'A'
    else:
        return 'A-'
    
```

```python
df['Grading'] = df['Total Marks'].apply(grading)
df
```

```python
def marking_system(df):
    a = df['Python Marks'] * 3
    b = df['Algorithm Marks'] * 2
    c = df['Data Structure Marks'] * 2
    return a+b+c

df['Exceptional Marks'] = df.apply(marking_system, axis = 1)
```
