```python
import numpy as np

# 1d array
arr1 = np.array([1,2,3,4])
print(arr1.ndim)

# 2d array
arr2 = np.array([[1,2,34,5],
                 [2,50,60,2]])
arr2
print(arr2.ndim)

#3d array

arr3 = np.array([
    #1st floor
    [[1,2,34,5],
    [2,50,60,2]],

    #2nd floor
    [[1,2,34,5],
    [2,50,60,2]],

    #3rd floor
    [[1,2,34,5],
    [2,50,60,2]],

    #4th floor
    [[1,2,34,5],
    [2,50,60,2]]

])

print(arr3.ndim)
```

#### Array attributes

```python
#dimentions Attribute
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#shape attribute
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#size
print(arr1.size)
print(arr2.size)
print(arr3.size)


```
