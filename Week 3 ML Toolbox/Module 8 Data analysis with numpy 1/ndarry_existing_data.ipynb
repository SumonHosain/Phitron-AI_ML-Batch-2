```python
import numpy as np
```

### List

```python
lst = [10,20,30,40,5]
arr = np.array(lst,dtype = np.int32)
print(type(arr))
print(arr)
print(arr.dtype)
```

```python
mix_list = [10,20,'hello',23.4,True]
arr= np.array(mix_list)
print(arr)
print(type(arr))
print(arr.dtype)
```

```python
matrix = [
    [[1,2,34,5],
    [2,50,60,2]],

    #2nd floor
    [[1,2,34,5],
    [2,50,60,2]],

    #3rd floor
    [[1,2,34,5],
    [2,50,60,2]]
]

arr = np.array(matrix)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

#### tuple

```python
tpl = (1,3,3,4)
arr = np.array(tpl)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

#### Set

```python
set = {1,2,3}
arr = np.array(list(set))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

#### Dict

```python
dic = {
    'a':[1,2,3],
    'b' :[2,34,5],
    'c': [55,4,6],
    'd':[4,3,5]
}
keys = dic.keys()
values = dic.values()
item = dic.items()

#keys to ndarray
arr = np.array(list(keys))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#values to ndarray
arr = np.array(list(values))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
# items to ndarray\arr = np.array(list(values))
arr = np.array(list(item))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.ndim)
```
