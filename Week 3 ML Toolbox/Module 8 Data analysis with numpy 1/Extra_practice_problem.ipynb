```python
arr = list(map(int,input('Entre your Celcious').split()))
print(arr)

def celcious_farenhite(celcious):
    farenhights = [] 
    for i in celcious:
         f = i * 1.8 + 32
         farenhights.append(f)
    return farenhights

farenhight = celcious_farenhite(arr)
print(farenhight)

```
