```python
import numpy as np
```

```python
# 1d array
arr1 = list(range(1,10))
arr1= np.array(arr1)
print(arr1.ndim)

# 2d Array
arr2=np.array([[1,3,4]
               ,[1,2,3],
               [21,2,4],
               [1,2,5]])
print(arr2.ndim)

# 3d array
arr3 = np.array([
    #1st floor
    [[1,2,3,5],
    [2,5,6,2]],

    #2nd floor
    [[1,2,34,5],
    [2,5,67,2]],

    #3rd floor
    [[1,2,35,5],
    [2,5,65,26]]
    ])
print(arr3.ndim)
```

```python
#indexing
print(arr1)
print(arr1[7])
print(arr3)
print(arr3[1][0][2])
```

```python
print(arr1)
```

```python
#slicing for 1d array = arr[start:end:step]
arr1_mod = arr1[1:5]
print(arr1_mod)
```

```python
print(arr2)
```

```python
#slicing for 2d array = arr[row_start:row_end:row_step, col_start:col_end:col_step]
arr2_mod = arr2[0:3:2,0:2:1]
print(arr2_mod.ndim)
```
