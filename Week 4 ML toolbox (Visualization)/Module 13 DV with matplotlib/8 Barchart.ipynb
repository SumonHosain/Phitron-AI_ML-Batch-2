```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Student':['B','C','D','E','F','G','H'],
    'Gender' : ['Male','Female','Male','Female','Male','Female','Male'],
    'Study_Hours':[2,5,3,7,4,8,1],
    'Attendance':[60,88,70,95,80,98,55],
    'Grade':['Fail','Pass','Fail','Pass','Pass','Pass','Fail']

}
```

```python
df = pd.DataFrame(data)
df
```

```python
gender_group = df.groupby('Gender').size()
gender_group
```

```python
plt.bar(gender_group.index,gender_group.values,edgecolor = 'Black', color = ['Red','Orange'])
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
df_2 = pd.read_csv('student_data.csv')
status = df_2.groupby('CompletionStatus').size()
plt.bar(status.index,status.values,edgecolor = 'Black', color = ['Green','Yellow','Red'])
plt.xlabel('Gender')
plt.ylabel('Count')
```
