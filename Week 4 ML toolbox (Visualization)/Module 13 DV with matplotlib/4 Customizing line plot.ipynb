```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('Enrollment Ratio')
plt.title('Student Enroll Data')
plt.plot(df['Year'],df['Programming'], label = 'Programing',color = 'red',linewidth = 3,linestyle = 'dashed', marker = 'o',markersize= 10)
plt.plot(df['Year'],df['Digital Marketing'],label = 'Marketing', color = 'Black',linewidth = 3)
plt.legend()
```
