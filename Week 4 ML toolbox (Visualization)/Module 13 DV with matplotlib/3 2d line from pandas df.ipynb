```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('Enrollment Ratio')
plt.title('Student Enroll Data')
plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
# adding lebel with legend
plt.xlabel('Year')
plt.ylabel('Enrollment Ratio')
plt.title('Student Enroll Data')
plt.plot(df['Year'],df['Programming'],label = 'Programing')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital marketing')
plt.legend(loc='lower right')
```
