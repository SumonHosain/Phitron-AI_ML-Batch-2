```python
import numpy as np
```

```python
#np.arange(start,end, step)
arr = np.arange(1,10,1).reshape(3,3)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#np.linespace(start,end, how_much_point_to_pick)
arr = np.linspace(1,10,12)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#np.logspace(start,end(included), how_much_point_to_pick)
arr = np.logspace(1,4,12)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```
