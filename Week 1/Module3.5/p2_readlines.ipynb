```python
# enumerate(iterable, start=0)

def read_lines(file_name):
    with open(file_name,'a+') as file:
        file.seek(0)
        for index_number, file in enumerate(file,start= 1):
            print(f'{index_number}:{file.strip()}')
        
read_lines('filename.txt')
```
