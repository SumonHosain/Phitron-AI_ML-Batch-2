```python
#Generator

lst = [x for x in range(0,500)]

def generator(chunk_size, lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]


x = generator(5,lst)
print(next(x))
print(next(x))
```

```python
#iterator

s = {1,2,3,4}
s_iter = iter(s)
print(next(s_iter))
print(next(s_iter))
```
