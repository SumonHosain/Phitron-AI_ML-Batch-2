Get the text from sample.txt , and count the following

Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

```python
from functools import reduce
with open('sample.txt') as file:
    strings = file.readlines()
    
    #total lines
    total_lines = len(strings)
    
    # total words
    '''def total_words(x):
        for i in x:
            len_word = len(i.split())
            print(len_word)
    
    hel = total_words(strings)'''
    
    #total words    
    total_words = list(map(lambda x:len(x.split()),strings))
    print(total_words)
    # total charecter
    total_charecter = reduce(lambda x,y:x+y,total_words)
    
    #no new line at the stings
    strings = list(map(str.strip,strings))
   
    
    #no space at the strings
    strings = list(map(lambda x:x.replace(" ",''),strings))
    
    total_charecter = list(map(lambda x:len(x),strings))
    total_charecter = len(reduce(lambda x,y:x+y, strings))
    

    with open('counter_of_stirngs.txt','w') as file_2:
        file_2.write(f' Total number of words: {total_words}\n Total number of lines: {total_lines}\n Total number of charecter:{total_charecter}')

```
