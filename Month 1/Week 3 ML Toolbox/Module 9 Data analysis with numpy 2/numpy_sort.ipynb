```python
import numpy as np
```

```python
#inplace sort er jonno default array er value change hobe 
a = np.array([1,3,5,10,0])
print(a)
a.sort() 
print(a)
```

```python
#copy sort er jonno default array er value change hobe na
x = np.array([3,1,10,15,12])
arr_sort = np.sort(x)
print(arr_sort)
print(x)
```

```python
# 2d array sorting
mat =np.array(
    [[8,4,16],
    [12,11,15]]
)

hor_sort = np.sort(mat,axis =1)
print(hor_sort)

var_sort = np.sort(mat,axis = 0)
print(var_sort)
```
