```python
a = np.random.randint(1,10,size=(2,3))
b = np.random.randint(20,30,size = (2,3))
print(a)
print(b)
```

```python
#concatenate
con_row = np.concatenate((a,b),axis = 0) #row_wise
print(con_row)

con_col = np.concatenate((a,b),axis =1) #col_wise 
print(con_col)
```

### Error during concatenate
row wise concate korar somoy col same thakbe, 
col wise concate korar somoy row same thakbe 

```python
# col error 
x = np.random.randint(1,10,size=(2,3))
y = np.random.randint(20,30,size = (1,3))

con_row = np.concatenate((x,y),axis = 0)
print(con_row)
```

```python
#col_error
con_col = np.concatenate((x,y), axis=1)
print(con_col)
```
