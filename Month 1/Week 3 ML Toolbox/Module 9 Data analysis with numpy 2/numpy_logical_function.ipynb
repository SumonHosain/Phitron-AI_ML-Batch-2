```python
import numpy as np
```

```python
# comaprison operator - >,< , >=,<= , ==, !=
a = np.array([1,22,55,33])
b = np.array([22,32,55,44])

greater_than = a>b
print(greater_than)

equal = a==b
print(equal) 

# all() -> sob gulo value true kina,
# any() -> jekono akta value true kina

print(np.all(greater_than))
print(np.any(equal))
```
