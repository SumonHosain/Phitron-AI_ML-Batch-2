```python
# need to explore mathmetical function of numpy
import numpy as np
```

```python
a = np.array(range(1,10))
b = np.array(range(2,11))
print(a)
print(b)
```

```python
add = a+b
print(add)

sub = a-b
print(sub)

mul = a*b
print(mul)

div = a/b
print(div)

remainder = a%b
print(remainder)
```

#### Mathmetical function in numpy

#### Trigonmetry

```python
sin_val = np.sin(a)
print(sin_val)

cos_val = np.cos(a)
print(cos_val)

#radian to degree
deg_con = np.rad2deg(a)
print(deg_con)

#deg to radian
red_con = np.deg2rad(deg_con)
print(red_con)
```

```python
#log
base_10_log_val = np.log10(a)  # kono number ke 10 diye kotobar vhag kora jabe seta log10 er result
print(base_10_log_val)

two_diye_vhag = np.log2(a)    # kono number ke 2 diye kotobar vhag kora jabe seta
print(two_diye_vhag)
```

```python
# squreroot
x = np.array([4,2,6,16,25])
sqrt = np.sqrt(x)
print(sqrt)
```

```python
# different 
sum = np.sum(x)
print(sum)

cumalative_sum = np.cumulative_sum(x)
print(cumalative_sum)
```
