```python
import numpy as np
```

```python
#np.zeros
arr = np.zeros((3,1,3))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
arr3 = np.array([
    #1st floor
    [[1,2,34,5],
    [2,50,60,2]],

    #2nd floor
    [[1,2,34,5],
    [2,50,60,2]],
    ])
```

```python
#np.zeros_like
arr = np.zeros_like(arr3)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
# np.like
arr = np.ones((1,2,3))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
# np.like
arr = np.ones_like(arr3)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#np.empty
arr = np.empty((4,2,2),dtype = np.int32)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#np.full

arr = np.full((4,3),10)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
arr = np.full((4,3),np.inf)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```
