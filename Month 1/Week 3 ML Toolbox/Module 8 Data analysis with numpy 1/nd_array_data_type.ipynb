```python
import numpy as np
```

#### nd array data type

```python
arr = np.array([1,2,3,4])
print(arr.dtype)

#upcasted to float
arr = np.array([1.2, 5, 5.6])
print(arr.dtype)

#upcasted to string
arr = np.array(['5',4,3])
print(arr.dtype)
```

#### Selecting a dtype for a array

```python
arr = np.array([1,2,300],dtype = np.int16) #defining data type
print(arr.dtype)


arr = np.array([1,2,300],dtype = np.int16)
arr.astype(np.int32)
```
