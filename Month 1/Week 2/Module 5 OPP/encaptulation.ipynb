```python
class Mobile:
    def __init__(self,name,model,imei):
        self.__name = name
        self.model = model
        self.__imei = imei #private

    def get_imei(self):
        return self.__imei
    
    def get_name(self):
        return self.__name
    
    def name_setter(self,name):
        self.__name = name

iphone = Mobile('phone','a5','axoafih')

print(iphone.get_imei())
iphone.__imei = '123'
print(iphone.get_imei())

iphone.get_name()
iphone.name_setter('Hphone')
iphone.get_name()
```
