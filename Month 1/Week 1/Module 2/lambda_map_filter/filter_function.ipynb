```python
#filter(function,iterable)
x = [1,2,34,5,5,6]

result = list(filter(lambda x:x % 2==1,x))
print(result)
```

```python
a = [45,60,39,59,50,53]
filtered = list(filter(lambda x: x>=50,a))
print(filtered)
```

```python
a = ['Sumon','Rana','henchi','asad','Sumon']
filtered = list(filter(lambda x: len(x)>=5,a))
print(filtered)
```
