```python
lst = [i for i in range(100)]
def batch_generator(lst,batch_size):
    for x in range(0,len(lst),batch_size):
        yield lst[x:x+batch_size]

n = batch_generator(lst,3)
print(next(n))
print(next(n))
print(next(n))

```
