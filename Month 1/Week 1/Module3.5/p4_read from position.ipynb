```python
data_1 = 'sample.txt'

def find_word(data_1):

  with open(data_1, 'r') as file:
    total_lines = file.readlines()

    words = []
    for line in total_lines:
      words.extend(line.lower().strip().split())

    frequency = {}
    for word in words:
      frequency[word] = frequency.get(word, 0) + 1

    max_word = max(frequency, key=frequency.get)

    print(max_word)

find_word(data_1)
```

```python
def length_word(sentense):
    
    sentense = sentense.strip().split()
    lenth = list(map(lambda x:len(x),sentense))
    print(lenth)


length_word('machine learning is fun')
```

```python
def mirror_number(number):
  number = str(number)
  number_2 = str(number)[::-1]
  
  if number == number_2:
    print('Mirror Number')
  else:
    print('Not mirror number')
  
  
mirror_number(994)
```
