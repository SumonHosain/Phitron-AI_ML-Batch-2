```python
# creating a new file and writing there
with open('test.txt','w') as file:
    file.write('Hello world')
```

```python
# writing on a existing file 
# its overwrites the content
with open('test.txt','w')as file:
    file.write('How is going learing\n')
```

```python
# no overwriting just append the text on later
# append mode
with open('test.txt','a') as file:
    file.write('No more')
```
