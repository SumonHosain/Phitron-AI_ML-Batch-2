```python
import pandas as pd
```

```python
data = {
    'Name':['Sumon','Rana','Akij','Sumon','Rana'],
    'City':['Satkhira','khulna','Motijhil Moddhopara Dhaka','Satkhira','khulna'],
    'Department': ['Hr','President','Company','hel','President'],
    'Saalry':[1000,200,24444,20,200]
}
df = pd.DataFrame(data)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df
```

```python
# permanently delete duplicates
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# droping duplicates based on col

df.drop_duplicates(subset='City')
```

```python
# keeping last duplicate
df.drop_duplicates(subset = 'City',keep= 'last')
```

```python
# keeping fast duplicate
df.drop_duplicates(subset = 'City',keep= 'first')
```
