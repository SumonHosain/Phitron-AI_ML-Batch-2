```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student_IQdata.csv')
df
```

```python
# relation between study hours and iq score -> positive corelation
plt.xlabel('Study Hour')
plt.ylabel('IQ Score')
plt.title('relation between study hours and iq score')
plt.scatter(df['Study_Hour'],df['IQ_Score'])
```

```python
# Relation between shoe size and IQ score -> No relation
plt.xlabel('shoe size')
plt.ylabel('Iq score')
plt.title('Relation between shoe size and IQ score')
plt.scatter(df['Shoe_Size'],df['IQ_Score'])
```

```python
# relation between chilling hours and iq socre -> Negative coo-relation
plt.xlabel('Chilling hours')
plt.ylabel('IQ Score')
plt.title('Relation between chillng hours and IQ Score')
plt.scatter(df['Chilling_Hours'],df['IQ_Score'])
```
